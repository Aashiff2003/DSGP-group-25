```python
#chech dataset structure
import os

dataset_path = "/content/drive/MyDrive/dataset 3"

# List classes (subdirectories)
classes = os.listdir(dataset_path)
print("Classes:", classes)

# Check number of images per class
for cls in classes:
    class_path = os.path.join(dataset_path, cls)
    if os.path.isdir(class_path):
        print(f"Class {cls}: {len(os.listdir(class_path))} images")
```

**Sample images**

```python
#display sample images
import cv2
import matplotlib.pyplot as plt
import random
import glob

def show_sample_images(dataset_path, classes, num_images=5):
    num_rows = len(classes)
    num_cols = num_images

    # Create figure with the right number of rows
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 10))

    for i, cls in enumerate(classes):
        img_files = glob.glob(os.path.join(dataset_path, cls, "*.jpg"))
        sample_files = random.sample(img_files, min(num_images, len(img_files)))

        for j in range(num_cols):
            if j < len(sample_files):
                img_path = sample_files[j]
                img = cv2.imread(img_path)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                axes[i, j].imshow(img)
                axes[i, j].axis("off")
            else:
                # Hide empty subplots
                axes[i, j].axis("off")

        # Set the title in the middle of each row
        axes[i, num_cols // 2].set_title(cls, fontsize=14, pad=10)

    plt.tight_layout()
    plt.show()

show_sample_images(dataset_path, classes)
```

**Checking aspect ratios and image sizes**

```python
#check image sizes
import numpy as np

image_sizes = []

for cls in classes:
    img_files = glob.glob(os.path.join(dataset_path, cls, "*.jpg"))
    for img_path in img_files:
        img = cv2.imread(img_path)
        if img is not None:
            image_sizes.append(img.shape[:2])  # (Height, Width)

image_sizes = np.array(image_sizes)
print("Mean Size:", np.mean(image_sizes, axis=0))
print("Min Size:", np.min(image_sizes, axis=0))
print("Max Size:", np.max(image_sizes, axis=0))
```

**Check for duplicates**

```python
#duplicates
import hashlib

def get_image_hash(image_path):
    with open(image_path, "rb") as f:
        return hashlib.md5(f.read()).hexdigest()

hashes = {}
duplicates = []

for cls in classes:
    img_files = glob.glob(os.path.join(dataset_path, cls, "*.jpg"))
    for img_path in img_files:
        img_hash = get_image_hash(img_path)
        if img_hash in hashes:
            duplicates.append(img_path)
        else:
            hashes[img_hash] = img_path

print(f"Found {len(duplicates)} duplicate images.")
```

**Check class distribution**

```python
import seaborn as sns
import pandas as pd

class_counts = {cls: len(glob.glob(os.path.join(dataset_path, cls, "*.jpg"))) for cls in classes}
df = pd.DataFrame(list(class_counts.items()), columns=["Class", "Count"])

plt.figure(figsize=(10, 5))
sns.barplot(x="Class", y="Count", data=df)
plt.xticks(rotation=45)
plt.title("Class Distribution")
plt.show()
```

**Color Analysis**

```python
import random
import glob
import os

def plot_color_histogram(image_path):
    img = cv2.imread(image_path)
    colors = ('b', 'g', 'r')
    plt.figure(figsize=(6, 4))

    for i, col in enumerate(colors):
        hist = cv2.calcHist([img], [i], None, [256], [0, 256])
        plt.plot(hist, color=col)

    plt.title("Color Histogram")
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")
    plt.show()
# Choose a random class
random_class = random.choice(classes)

# Get a list of images in the random class
img_files = glob.glob(os.path.join(dataset_path, random_class, "*.jpg"))
random_image = random.choice(img_files)
plot_color_histogram(random_image)
```
